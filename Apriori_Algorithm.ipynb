```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('/content/groceries.csv')
```

```python
dataset.head()
```

```python
pd.set_option('display.max_rows',500)
```

Make a list of list for items


```python
market=[]
for i in range(0,dataset.shape[0]):
  cus=[]
  for j in dataset.columns:
    if type(dataset[j][i])==str:
      cus.append(dataset[j][i])
  market.append(cus)
```

```python
l=[]
for i in market:
  for j in i:
    l.append(j)
```

```python
import collections
```

```python
p=collections.Counter(l)
```

```python
p.keys()
```

```python
p.values()
```

```python
d={'item_name':p.keys(),'values':p.values()}
```

```python
pd.DataFrame(d).sort_values(by=['values'],ascending=False)
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
tr=TransactionEncoder()
```

```python
tr.fit(market)
```

```python
df=pd.DataFrame(tr.transform(market),columns=tr.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df,min_support=0.05,use_colnames=True).sort_values(by=['support'],ascending=True)
```
